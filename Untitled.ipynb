```python
def p1(N0,N1,circ):
    n0= N0
    n1= N1
    s= 0
    H= 0
    for i in range(len(circ)):
        if circ[i] is 'x':
            temp= n0
            n0= n1
            n1= temp
            s= (s+1)%2
        elif circ[i] is 's':
            n1= (n1 + 8)%16
        elif circ[i] is 't':
            n1= (n1 + 2)%16
        elif circ[i] is 'h':
            if H is 0 and s is 0:
                n1= n0
                H= 1
            if H is 0 and s is 1:
                n0= n1
                n1= (n1 + 8)%16
                H= 1
            if H is 1:
                n0= ((n0 + n1))/2
                n1= ((n0 + n1 + 8))/2
    return 16*n1 + n0
```
