```python
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, Aer, execute
import math as m
```

```python
def r1(r0,n0,n1,circ):
    q = QuantumRegister(1)
    c = ClassicalRegister(1)
    qc = QuantumCircuit(q, c)
    
    qc.u3(2*m.acos(r0),n1*m.pi/8,0,q)
    qc.x(q)
    qc.u1(n0*m.pi/8,q)
    qc.x(q)
    for i in range(len(circ)):
        if circ[i] is 'x':
            qc.x(q)
        elif circ[i] is 's':
            qc.s(q)
        elif circ[i] is 't':
            qc.t(q)
        elif circ[i] is 'h':
            qc.h(q)
    
    qc.measure(q, c)
    
    backend = Aer.get_backend('qasm_simulator')
    job_sim = execute(qc, backend)
    sim_result = job_sim.result()

    return sim_result.get_counts(qc)
```

```python
r1(1,0,0,'stsxhm')
```

```python
def p1(N0,N1,circ):
    n0= N0
    n1= N1
    s= 0
    H= 0
    for i in range(len(circ)):
        if circ[i] is 'x':
            temp= n0
            n0= n1
            n1= temp
            s= (s+1)%2
        elif circ[i] is 's':
            n1= (n1 + 8)%16
        elif circ[i] is 't':
            n1= (n1 + 2)%16
        elif circ[i] is 'h':
            if H is 0 and s is 0:
                n1= n0
                H= 1
            if H is 0 and s is 1:
                n0= n1
                n1= (n1 + 8)%16
                H= 1
            if H is 1:
                n0= ((n0 + n1))/2
                n1= ((n0 + n1 + 8))/2
        elif circ[i] is 'm':
            #Do nothing
    return 16*n1 + n0
```

```python
p1(0,0,'xxssh')
```

```python
def eval(r0,n0,n1,q):
    R0= r1(r0,n0,n1,q)['0']/1024
    N0= p1(n0,n1,q)%16
    N1= ( p1(n0,n1,q) - N0)/16
    
    res= [R0,N0,N1]
    return res
```

```python
eval(1,0,0,'xxxsh')
```

```python
def exec(C,R,N0,N1):
    i= 0
    
    res= []
    for line in C:
        res.append(eval(R[i],N0[i],N1[i],line))
    return res
```

```python
exec(['xx','xsht','ssht','xxshtsst'],[0.2,0.3,0.5,0.9],[1,2,1,1],[1,2,1,2])
```

```python
def cont(R,N0,N1,C1,C2,CR,CN0,CN1):
    
    Q= []
    Q1= []
    Q2= []
    for I in exec(C1,R,N0,N1):
        I[0]= I[0]*CR
        I[1]= ( I[1] + CN0 )%16
        I[2]= ( I[2] + CN0 )%16
        Q1.append(I)
    for I in exec(C2,R,N0,N1):
        I[0]= I[0]*(1-CR)
        I[1]= ( I[1] + CN1 )%16
        I[2]= ( I[2] + CN1 )%16
        Q2.append(I)
```
