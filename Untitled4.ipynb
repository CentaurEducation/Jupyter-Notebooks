```python
mem= []
for i in range(1000):
    mem.append([1,0,0])
```

```python
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, Aer, execute
import math as m

def r(r0,n0,n1,circ):
    q = QuantumRegister(1)
    c = ClassicalRegister(1)
    qc = QuantumCircuit(q, c)
    
    qc.u3(2*m.acos(r0),n1*m.pi/8,0,q)
    qc.x(q)
    qc.u1(n0*m.pi/8,q)
    qc.x(q)
    for i in range(len(circ)):
        if circ[i] is 'x':
            qc.x(q)
        elif circ[i] is 's':
            qc.s(q)
        elif circ[i] is 't':
            qc.t(q)
        elif circ[i] is 'h':
            qc.h(q)
    
    qc.measure(q, c)
    
    backend = Aer.get_backend('qasm_simulator')
    job_sim = execute(qc, backend)
    sim_result = job_sim.result()

    if '0' in sim_result.get_counts(qc):
        return sim_result.get_counts(qc)['0']/1024
    else:
        return sim_result.get_counts(qc)['1']/1024

def p0(N0,N1,circ):
    n0= N0
    n1= N1
    s= 0
    H= 0
    for i in range(len(circ)):
        if circ[i] is 'x':
            temp= n0
            n0= n1
            n1= temp
            s= (s+1)%2
        elif circ[i] is 's':
            n1= (n1 + 8)%16
        elif circ[i] is 't':
            n1= (n1 + 2)%16
        elif circ[i] is 'h':
            if H is 0 and s is 0:
                n1= n0
                H= 1
            if H is 0 and s is 1:
                n0= n1
                n1= (n1 + 8)%16
                H= 1
            if H is 1:
                n0= ((n0 + n1))/2
                n1= ((n0 + n1 + 8))/2
    return n0

def p1(N0,N1,circ):
    n0= N0
    n1= N1
    s= 0
    H= 0
    for i in range(len(circ)):
        if circ[i] is 'x':
            temp= n0
            n0= n1
            n1= temp
            s= (s+1)%2
        elif circ[i] is 's':
            n1= (n1 + 8)%16
        elif circ[i] is 't':
            n1= (n1 + 2)%16
        elif circ[i] is 'h':
            if H is 0 and s is 0:
                n1= n0
                H= 1
            if H is 0 and s is 1:
                n0= n1
                n1= (n1 + 8)%16
                H= 1
            if H is 1:
                n0= ((n0 + n1))/2
                n1= ((n0 + n1 + 8))/2
    return n1

def ctrl_u( ctr, tgt, U ):
    
    R1= mem[ctr][0]
    R2= mem[tgt][0]
    n0= mem[tgt][1]
    n1= mem[tgt][2]
    
    #Prep temporary qubit
    q = QuantumRegister(1)
    c = ClassicalRegister(1)
    qc = QuantumCircuit(q, c)
    
    qc.u3(2*m.acos(R2),n1*m.pi/8,0,q)
    qc.x(q)
    qc.u1(n0*m.pi/8,q)
    qc.x(q)
    
    A= r( R2, n0, n1, U )
    
    mem[tgt][0]= A*(1 - R2) + R2*R1
    mem[tgt][1]= 0
    mem[tgt][2]= 0
```

```python
def apply( addr, U ):
    mem[addr][0]= r( mem[addr][0], mem[addr][1], mem[addr][2], U )
    mem[addr][1]= p0( mem[addr][1], mem[addr][2], U )
    mem[addr][2]= p1( mem[addr][1], mem[addr][2], U )
```

```python
apply(8,'xshtx')
```
